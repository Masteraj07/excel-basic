1- Explain the difference between Relative , Absolute and Mixed Cell Referencing.
# You have the following dataset:
  Ans- 1:- Relative Cell Referencing

Relative cell referencing changes automatically when a formula is copied to another cell.

Using the given data set:
If we calculate 10% tax for the Car in cell C2 using:
=B2*10%
When this formula is copied down:

For Bicycle → =B3*10%

For Auto-Rickshaw → =B4*10%

Here, the cell reference changes from B2 to B3 and B4, showing relative referencing.

2:- Absolute Cell Referencing

Absolute cell referencing keeps a cell fixed, even when the formula is copied.
It uses the $ symbol.

Using the given data set:
Assume 10% tax rate is stored in D1.

Formula in C2: =B2*$D$1

When copied down for Bicycle and Auto-Rickshaw:
C3 → =B3*$D$1

C4 → =B4*$D$1

The cell $D$1 remains unchanged, which demonstrates absolute referencing.

3- Mixed Cell Referencing

Mixed cell referencing fixes either the row or the column, but not both.

Using the given data set:
Formula: =$B2*C$1

$B → Column B (Price) is fixed

2 → Row changes for Bicycle and Auto-Rickshaw

$1 → Row 1 remains fixed

This type of reference is useful when working with structured tables and repeated calculations.
